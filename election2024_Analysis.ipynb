```python
import pandas as pd

# Load the dataset (adjust the file path as necessary)
file_path = r'election2024.csv'
df = pd.read_csv(file_path)

# Display the first few rows to confirm the data is loaded correctly
df.head()
```

```python
# Get an overview of the dataset to understand its structure and contents.
# Basic information about the dataset
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python

# Check for missing values
df.isnull().sum()
```

## Perform Correlation Analysis:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize = (12,8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', linewidths = 0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Income Distribution:

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'income', data = df, order = df['income'].value_counts().index)
plt.title('Income Distribution')
plt.show()
```

## Education Distribution:

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='education', data=df, order=df['education'].value_counts().index)
plt.title('Education Distribution')
plt.show()
```

## Employment Status Distribution:

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='employment', data=df, order=df['employment'].value_counts().index)
plt.title('Employment Status Distribution')
plt.show()
```

## Job Type Distribution:

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='job_type', data=df, order=df['job_type'].value_counts().index)
plt.title('Job Type Distribution')
plt.show()
```

## Regional Voting Patterns:

### Voting Intentions by Federal District

```python
plt.figure(figsize=(15, 10))
sns.countplot(y='federal_district', hue='will_vote', data=df, order=df['federal_district'].value_counts().index)
plt.title('Voting Intentions by Federal District')
plt.show()
```

### Candidate Preferences by Federal District

```python
plt.figure(figsize=(15, 10))
sns.countplot(y='federal_district', hue='candidate', data=df, order=df['federal_district'].value_counts().index)
plt.title('Candidate Preferences by Federal District')
plt.show()
```
